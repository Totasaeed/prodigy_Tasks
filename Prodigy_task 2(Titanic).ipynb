# <span style="color:white">Import Libraries  </span>

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

# <span style="color:white">Loading Data  </span>

```python
df = pd.read_csv(r"E:\سنه تالتة تيرم اول\Python\pythonProject\train.csv")
df.head()
```

# <span style="color:white">Data Preparation  </span>

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.describe(include=['O'])
```

```python
df.isnull().sum()
```

```python
for col in df.columns:
    if df[col].isnull().sum() > 0 and df[col].dtype == 'O':  
        df[col].fillna(df[col].mode().iloc[0], inplace=True)  
    elif df[col].isnull().sum() > 0 and df[col].dtype != 'O':  
        df[col].fillna(df[col].mean(), inplace=True)  

df.isnull().sum()
```

# <span style="color:White">Exploratory Data Analysis </span> 

```python
plt.pie(df['Survived'].value_counts(),labels=['Unurvived', 'Survived'],
        autopct='%1.1f%%', startangle=90,
         colors=['skyblue', 'blue'],
         explode = (0.1, 0))

plt.axis('equal') 
plt.style.use('dark_background') 
plt.show()
```

```python
for col in ['Sex','Pclass']:
        df[col].value_counts()
        plt.figure(figsize=(8, 6))
        sns.countplot(x=col, hue='Survived', data=df,color='skyblue',alpha=.9)
        plt.title(f'Count plot for {col}')
        plt.style.use('dark_background')
        plt.show()
```

# <span style="color:White"> Encoding & Scaling </span>

```python

for col in ['Sex','Embarked']:
        encoder = LabelEncoder()
        df[col] = encoder.fit_transform(df[col])

for col in ['Age','Fare']:     
    scaler = StandardScaler()
    df[col] = scaler.fit_transform(df[col].values.reshape(-1,1))
        
df.head()
```

```python
for col in df.columns:
    if df[col].dtype == "object":
        df.drop(col, axis=1, inplace=True)
        
df.head()
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(), annot=True, cmap = "mako")
```

# <span style="color:white">Splitting the data into training data & Test data</span>

```python
X = df.drop(columns=['PassengerId','Survived'],axis=1)
Y = df["Survived"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
```

# <span style="color:white">Model Building : Logistic Regression</span>

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

# <span style="color:white">Model Evaluation</span>

```python
X_train_prediction = model.predict(X_train)
print(X_train_prediction)
```

```python
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('Accuracy score of training data : ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
print(X_test_prediction)
```

```python
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('Accuracy score of test data : ', test_data_accuracy)
```

```python
print(f"Training accuracy : {model.score(X_train, Y_train)}")
print(f"Test accuracy {model.score(X_test, Y_test)}")
```

# <span style="color:white">Confusion Matrix</span>

```python
conf_matrix = confusion_matrix(Y_test, X_test_prediction)

plt.figure(figsize=(8, 6))
sns.set(font_scale=1.2)  
sns.heatmap(conf_matrix, annot=True, fmt="d", linewidths=0.5, cbar=False, cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')

plt.show()
```
